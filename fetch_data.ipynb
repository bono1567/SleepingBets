# ⚽ Sofascore Data Fetcher — 2025/26 Season
Fetches player and team stats for all 5 top European leagues and caches them as CSV files in `./data/`.

```python
# !pip install ScraperFC pandas
```

```python
import ScraperFC as sfc
import pandas as pd
import os
import time

YEAR = "25/26"
DATA_DIR = "./data"
os.makedirs(DATA_DIR, exist_ok=True)

LEAGUES = [
    "England Premier League",
    "Spain La Liga",
    "Germany Bundesliga",
    "France Ligue 1",
    "Italy Serie A",
]

ss = sfc.Sofascore()
print("Sofascore client ready.")
```

## 1. Fetch Player League Stats (all accumulations)

```python
accumulations = ["total", "per90", "perMatch"]

for accum in accumulations:
    all_frames = []
    for league in LEAGUES:
        safe_name = league.replace(" ", "_")
        cache_file = f"{DATA_DIR}/players_{safe_name}_{accum}.csv"

        if os.path.exists(cache_file):
            print(f"  [CACHED] {league} ({accum})")
            df = pd.read_csv(cache_file)
        else:
            print(f"  [FETCHING] {league} ({accum}) ...")
            try:
                df = ss.scrape_player_league_stats(YEAR, league, accumulation=accum)
                df["league"] = league
                df.to_csv(cache_file, index=False)
                print(f"    Saved {len(df)} rows → {cache_file}")
                time.sleep(2)  # be polite to the API
            except Exception as e:
                print(f"    ERROR: {e}")
                continue

        all_frames.append(df)

    if all_frames:
        combined = pd.concat(all_frames, ignore_index=True)
        combined_path = f"{DATA_DIR}/players_all_{accum}.csv"
        combined.to_csv(combined_path, index=False)
        print(f"\n✅ Combined {accum} → {combined_path} ({len(combined)} rows)\n")
```

## 2. Fetch Team League Stats

```python
all_team_frames = []

for league in LEAGUES:
    safe_name = league.replace(" ", "_")
    cache_file = f"{DATA_DIR}/teams_{safe_name}.csv"

    if os.path.exists(cache_file):
        print(f"  [CACHED] {league} teams")
        df = pd.read_csv(cache_file)
    else:
        print(f"  [FETCHING] {league} teams ...")
        try:
            df = ss.scrape_team_league_stats(YEAR, league)
            df["league"] = league
            df.to_csv(cache_file, index=False)
            print(f"    Saved {len(df)} rows → {cache_file}")
            time.sleep(2)
        except Exception as e:
            print(f"    ERROR: {e}")
            continue

    all_team_frames.append(df)

if all_team_frames:
    combined_teams = pd.concat(all_team_frames, ignore_index=True)
    combined_teams.to_csv(f"{DATA_DIR}/teams_all.csv", index=False)
    print(f"\n✅ Combined teams → {DATA_DIR}/teams_all.csv ({len(combined_teams)} rows)")
```
